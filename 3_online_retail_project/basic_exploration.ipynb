```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df1= pd.read_excel("C:/Users/prash/Desktop/data/online_retail_II.xlsx", sheet_name='Year 2009-2010')
df2= pd.read_excel('C:/Users/prash/Desktop/data/online_retail_II.xlsx', sheet_name='Year 2010-2011')
```

```python
# Combine df1 and df2 into a single DataFrame
df = pd.concat([df1, df2], ignore_index=True)

# Display the combined DataFrame
df
```

```python
df.info()
```

```python
df['StockCode'].value_counts()
```

```python
df[df['Description'].isnull()]['StockCode'].unique()
```

```python
# Assuming 'InvoiceDate' is the datetime column in your DataFrame
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day

# Display the updated DataFrame
df.head()
```

```python
df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
df['Description'].value_counts().head(10)
```

```python
# Assuming 'InvoiceDate' is the datetime column in your DataFrame
df['Quarter'] = df['InvoiceDate'].dt.quarter

# Display the updated DataFrame
df.head()
```

```python
# Assuming 'Month' is the column with month numbers
df['MonthName'] = pd.to_datetime(df['Month'], format='%m').dt.month_name()

# Display the updated DataFrame
df.head()
```

```python
df['sales'] = df['Quantity'] * df['Price']
df.sample(10)
```

```python
df2010sales=df[df['Year']==2010].groupby(['Month','MonthName']).agg({'sales':'sum'})
df2010sales
```

```python
df2010sales.reset_index(inplace=True)
df2010sales
```

```python
df2010sales.drop(columns='Month', inplace=True)
df2010sales
```

```python
sns.barplot(x='MonthName', y='sales', data=df2010sales, palette='viridis',hue='MonthName')
plt.title('Monthly Sales in 2010')
plt.xlabel('')
plt.ylabel('Sales')
sns.despine()
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
df2011sales=df[df['Year']==2011].groupby(['Month','MonthName']).agg({'sales':'sum'})
df2011sales
```

```python
df2011sales.reset_index(inplace=True)
df2011sales
```

```python
df2011sales.drop(columns='Month', inplace=True)
df2011sales
```

```python
sns.barplot(x='MonthName', y='sales', data=df2011sales, palette='viridis',hue='MonthName')
plt.title('Monthly Sales in 2011')
plt.xlabel('')
plt.ylabel('Sales')
sns.despine()
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming df2010sales and df2011sales are already prepared
# Merge the two DataFrames on 'MonthName' for comparison
comparison_df = pd.merge(df2010sales, df2011sales, on='MonthName', suffixes=('_2010', '_2011'))

# Plot the barplot for 2010 sales
sns.barplot(x='MonthName', y='sales_2010', data=comparison_df, color='skyblue', label='2010 Sales')

# Add a line plot for 2011 sales
plt.plot(comparison_df['MonthName'], comparison_df['sales_2011'], color='red', marker='o', label='2011 Sales')

# Add labels, title, and legend
plt.title('Month-on-Month Sales Comparison (2010 vs 2011)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.legend()
sns.despine()
plt.tight_layout()

# Show the plot
plt.show()
```
